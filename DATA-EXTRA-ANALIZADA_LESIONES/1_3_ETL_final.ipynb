## En esta parte del ETL vamos a concatenar los dos archivos generados anteriormente 


### <center>Importar Librerias</center>

```python
import pandas as pd
import numpy as np
import os
import funciones

import warnings
warnings.filterwarnings("ignore")
```

________________

### <center>Extracción de datos</center>

```python
#Se pasa dirección de los archivos
archivo_h = os.path.join('..', 'CSV', 'homicidio.csv')
archivo_l = os.path.join('..', 'CSV', 'lesiones.csv')

#Se extrae la información a sus respectivos dataframe
df_homicidio = pd.read_csv(archivo_h)
df_lesiones = pd.read_csv(archivo_l)
```

_________________________

### Generación de una nueva columna FATAL

Para poder unirlos y saber a cual dataframe pertenecia el registro, vamos a generar una nueva columna con valores 0 y 1 (es como darle valores Booleanos tipo True y False, si fue letal o no, homicidio = 1, lesiones = 0 ) 

```python
# Se agrega columna con valor = 1 en todos sus registros
df_homicidio['FATAL'] = 1
df_homicidio.head(2)
```

```python
df_homicidio.shape
```

```python
# Se agrega columna con valor = 1 en todos sus registros
df_lesiones['FATAL'] = 0
df_lesiones.head(2)
```

```python
df_lesiones.shape
```

__________________________

## Se concatena los dataframe ya que poseen las mismas columnas

```python
df_siniestros = pd.concat([df_homicidio, df_lesiones], ignore_index=True)
df_siniestros
```

______________

### Se verifican los datos

```python
funciones.verificar_tipo_datos(df_siniestros)
```

____________________

### <center>Se almacena en un archivos csv</center>

```python
# Se construye la ruta completa al archivo CSV desde la ubicación del notebook
siniestros = os.path.join('..', 'CSV', 'siniestros.csv')

# Se guarda el DataFrame en el archivo CSV
df_siniestros.to_csv(siniestros, index=False, encoding='utf-8')

# Se emite un mensaje que se guardó
print(f'Se guardó el archivo {siniestros}')
```
